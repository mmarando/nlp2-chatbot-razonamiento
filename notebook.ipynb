```python
def query_razonamiento(question):
    result = solve_complex_question(question)

    print("\n=== Razonamiento intermedio ===")
    for agent_name, reasoning in result["intermediate_reasoning"].items():
        print(f"\nüîπ {agent_name}:")
        print(reasoning)

    print("=== Respuesta final ===")
    print(result["answer"])

    print("\n=== Tokens utilizados ===")
    print(f"- Entrada (prompts): {result['tokens']['input']}")
    print(f"- Salida (respuestas): {result['tokens']['output']}")
    print(f"- Razonamiento (procesamiento interno): {result['tokens']['reasoning']}")
    print(f"- Total: {result['tokens']['total']}")
```
